```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("../sales.db")
df = pd.read_sql("SELECT * FROM sales", conn)
df.head()
```

```python
df["revenue"] = df["price"] * df["quantity"]
df.head()
```

```python
total_revenue = df["revenue"].sum()
print("Total Revenue:", total_revenue)
```

```python
category_rev = df.groupby("category")["revenue"].sum()
category_rev.plot(kind="bar", title="Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
top_products = df.groupby("product")["quantity"].sum().sort_values(ascending=False)
top_products.plot(kind="bar", title="Top Selling Products")
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.show()
```

```python
region_rev = df.groupby("region")["revenue"].sum()
region_rev.plot(kind="pie", autopct="%1.1f%%", title="Revenue by Region")
plt.show()
```

## Key Business Insights

- Electronics category generates the highest revenue.
- Laptop and Phone are top-selling products.
- North and West regions contribute the largest share of revenue.
- Accessories sell in higher volume but produce lower revenue.

## Conclusion
This analysis helps business teams optimize product focus,
pricing strategy, and regional marketing.
